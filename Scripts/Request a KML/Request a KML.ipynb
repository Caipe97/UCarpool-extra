```python
from geo2kml import to_kml
from urllib.request import urlopen
import json
```

```python
tripId = '10e4fe6b-5164-4f7e-a2be-b6cbc9d7f123'
link = 'http://172.23.215.114:3000/tripStats?tripId=%s'%(tripId)
f = urlopen(link)
myfile = f.read()
```

```python
jsonData = json.loads(myfile)


print(jsonData['realTimeData'][0]['timestamp'])
# Crear un GeoJSON
geoJSON = {
'type': 'Feature',
'properties': {},
'geometry': {},
'type': 'LineString',
'coordinates': [[position['coords']['longitude'], position['coords']['latitude']] for position in jsonData['realTimeData']]
}
kml_str = to_kml(geoJSON)

g = open("tripId-%s.kml"%(tripId), 'w');
g.write(kml_str)
```

## Ordenar por Timestamp en vez de orden de escritura

```python
sortedPings = sorted(jsonData['realTimeData'], key= lambda x: x['timestamp'])
# print(sortedPings)
# Crear un GeoJSON
geoJSON = {
'type': 'Feature',
'properties': {},
'geometry': {},
'type': 'LineString',
'coordinates': [[position['coords']['longitude'], position['coords']['latitude']] for position in sortedPings]
}
kml_str = to_kml(geoJSON)
print(kml_str)
g = open("sortedtripId-%s.kml"%(tripId), 'w');
g.write(kml_str)
```
