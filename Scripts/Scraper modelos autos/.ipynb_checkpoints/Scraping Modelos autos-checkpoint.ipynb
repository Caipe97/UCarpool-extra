```python
# Importar dependencias
import pandas as pd
import tabula as tb
import numpy as np
import re
import os
```

```python
# Importar archivo
file = 'valuacion.pdf'
df = tb.read_pdf(file, pages='all', area = (73.76, 105.8, 556.16, 243.59), columns = [106, 143.4, 196.33])
```

### Vamos bien, tenemos que agarrar todas las filas que sean NaN salvo Desc. Modelo y appendearlas al registro anterior

```python
mergedDf = pd.concat([data for data in df], ignore_index = True)
```

```python
mergedDf['Desc. Tipo'].unique()
```

```python
included = ['TODO TERRENO', 'FURGON', 'SEDAN 3 PUERTAS', 'SEDAN 4 PUERTAS',
       'SEDAN 5 PUERTAS', 'RURAL 5 PUERTAS', 'COUPE', 'DESCAPOTABLE',
       'SEDAN 2 PUERTAS', 'PICK-UP', 'TRANS.DE PASAJEROS', 'FAMILIAR',
       'CONVERTIBLE', 'UTILITARIO',
       'PICK-UP CABINA SIMPLE', 'PICK-UP CARROZADA',
       'PICK-UP CABINA DOBLE', 'RURAL 3 PUERTAS', 'FURGON O UTILITARIO',
       'FURGONETA', 'FURGONETA O UTILITARIO', 'PICK-UP CABINA Y MEDIA',
       'TRANSPORTE DE PASAJEROS', 'RURAL 4/5 PUERTAS', 'RURAL 4 PUERTAS', 'SEDAN 3 PUERTAS CON PORTON','JEEP','PICK UP']
df = mergedDf.drop(['Unnamed: 0'], axis=1)

dropIndexes = []

# Quitar filas con campos vacíos debido a formato de tabla

for i in range(1, df.shape[0]):
    if pd.isna(df['Desc. marca'][i]):
        if not pd.isna(df['Desc. Tipo'][i]):
            df['Desc. Tipo'][i-1] =  str(df['Desc. Tipo'][i-1]) + ' ' + str(df['Desc. Tipo'][i])
        if not pd.isna(df['Desc. Modelo'][i]):
            df['Desc. Modelo'][i-1] =  str(df['Desc. Modelo'][i-1]) + ' ' + str(df['Desc. Modelo'][i])
        dropIndexes.append(i)
df = df.drop(dropIndexes)
```

```python
# Incluir sólo las categorías que nos interesan
df = mergedDf[mergedDf['Desc. Tipo'].isin(included)]
```

```python
# Renombrar categorías a nombres más familiares
autos = ['SEDAN 3 PUERTAS', 'SEDAN 4 PUERTAS',
       'SEDAN 5 PUERTAS', 'RURAL 5 PUERTAS', 'COUPE', 'DESCAPOTABLE',
       'SEDAN 2 PUERTAS', 'CONVERTIBLE', 'RURAL 3 PUERTAS',]
camionetas = ['TODO TERRENO', 'PICK-UP', 'FAMILIAR', 'UTILITARIO', 'PICK-UP CABINA SIMPLE', 'PICK-UP CARROZADA',
       'PICK-UP CABINA DOBLE', 'RURAL 3 PUERTAS', 'FURGON O UTILITARIO',
       'FURGONETA', 'FURGONETA O UTILITARIO', 'PICK-UP CABINA Y MEDIA',
       'TRANSPORTE DE PASAJEROS', 'JEEP', 'PICK UP', 'RURAL 4/5 PUERTAS', 'RURAL 4 PUERTAS', 'SEDAN 3 PUERTAS CON PORTON',]
    
def simplifyNames(row):
    if row in autos:
        return 'AUTOMÓVIL'
    if row in camionetas:
        return 'CAMIONETA'
    return 'OTROS'

df['Desc. Tipo'] = df['Desc. Tipo'].apply(simplifyNames)
df = df.drop(['Unnamed: 0'], axis=1)
df = df[df['Desc. marca'].notna()]
df = df[df['Desc. Modelo'].notna()]
df = df.reset_index(drop=True)
```

```python
tipoVehiculosList = df['Desc. Tipo'].unique().tolist()
tipoVehiculosList.sort()
```

```python
# Crear Dataframe de Tipos de Vehículo

tipoVehiculosSchema = {
    'type': tipoVehiculosList
}
tipoVehiculos = pd.DataFrame(tipoVehiculosSchema)
```

```python
marcasList = df['Desc. marca'].unique().tolist()
marcasList = [str(i) for i in marcasList]
marcasList.sort()
```

```python
marcasSchema = {
    'make': marcasList
}
```

```python
marcaVehiculos = pd.DataFrame(marcasSchema)
```

#### Falta asignar Unique IDs a cada vehículo para su marca y tipo

```python
marcaVehiculos[marcaVehiculos['make'] == 'YAMAHA'].index.values[0]
```

```python

def getMarcaId(row):
    marca = marcaVehiculos[marcaVehiculos['make'] == row[0]].index.values[0] + 1
    return marca

def getTipoId(row):
    tipo = tipoVehiculos[tipoVehiculos['type'] == row[2]].index.values[0] + 1
    return tipo


df['makeId'] = df.apply(getMarcaId, axis=1)
df['typeId'] = df.apply(getTipoId, axis=1)

    
df = df.reset_index(drop=True)

df = df[['Desc. Modelo', 'makeId', 'typeId']]
df = df.rename(columns={"Desc. Modelo": "model"})
```

```python
df = df.drop_duplicates(subset = ['model', 'makeId', 'typeId'])
```

## Guardar archivos

```python
df.index = range(1, df.shape[0] + 1)
df.index.name = 'id'
df.to_csv('modelos.csv')

tipoVehiculos.index = range(1, tipoVehiculos.shape[0] + 1)
tipoVehiculos.index.name = 'id'
tipoVehiculos.to_csv('tipos.csv')
marcaVehiculos.index = range(1, marcaVehiculos.shape[0] + 1)
marcaVehiculos.index.name = 'id'
marcaVehiculos.to_csv('marcas.csv')
```
